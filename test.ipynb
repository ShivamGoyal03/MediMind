```python
import nltk
```

```python
# download nltk data in .venv directory as nltk_data folder
nltk.download('all', download_dir='.venv/nltk_data')
```

```python
# download raw data from (https://huggingface.co/datasets/ruslanmv/ai-medical-chatbot) 
# and save it in the data folder
import pandas as pd

ds = pd.read_parquet("hf://datasets/ruslanmv/ai-medical-chatbot/dialogues.parquet")
# Save the dataset to Parquet files in the 'data/raw' folder
import os

# Create 'data/raw' folder if it doesn't exist
os.makedirs('data/raw', exist_ok=True)

# Save each split (train, validation, test) to separate Parquet files
for split in ds.keys():
    output_file = f'data/raw/{split}.csv'
    ds[split].to_csv(output_file, index=False)
    print(f'Saved {len(ds[split])} examples to {output_file}')
```
